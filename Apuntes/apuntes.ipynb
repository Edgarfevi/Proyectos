### TLegend

La clase Tlegend dentro de ROOT una vez lo importamos tiene los siguientes argumentos

Leyenda = **Tlegend**(x1,y1,x2,y2)

Esto de momento unicamente nos define lo que ocupará en nuestro Canvas el cuadro de de la leyenda, el texto entonces se ajustará al tamaño que nostros estemos definiendo aquí.
Estas cordenadas a pesar de ser abstractacas definen los limites del cuadro, y estan normalizadas, es decir los posibles valores que pueden tomar estar variables van entre 0 a 1.
Es porcentaje del tamaño de nuestro Canvas siendo entonces 0 el borde izquierdo y 1 es el borde derecho para las x, para las y 0 es la parte inferior y 1 la parte superior.
De esta manera si entendemos como esta funcionando veremos que entonces lo que ocupa en x nuestro cuadro es $|x_1-x_2|$, siendo esto porcentaje de nuestro Canvas, por ejemplo escogiendo $x_1=0$, $x_2=0.2$ lo que ocupa segun lo que hemos dicho es:

- $|x_1-x_2| = |0-0.2| = 0.2$, por lo tanto ocupa un 20% de todo lo que mide de ancho nuestro Canvas.

Ademas podemos saber exactamente donde se situará, empezará a la izquierda del todo de nuestro canvas y se desplazará hacia la derecha hasta ocupar un 20% del Canvas.
Por lo tanto podemos generalizar el entenidimiento de la posicion del Tlengend como busquemos el valor más pequeño entre $x_1$ y $x_2$ ese valor será la distancia en porcentaje con respecto al borde izquierdo y se desplazará hacia la derecha ocupando el cuadro en totalidad el porcentaje de lo que definimos anteriormente.
No importa si $x_1$ es mayor o $x_2$, es indistinto, por eso no menciono en el proceso de entender donde esta dispuesto que x1 tenga que ser menor o mayor. Retomando el ejemplo anterior encontrariamos la misma disposición si $x_1=0.2$, $x_2=0$
