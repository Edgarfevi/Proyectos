```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
with open("fondo_numeros", "r") as file:
    data = file.read()
    data = data.replace(',', '.')

datos = data.split('\n')
datos_buenos = [datos.split('\t') for datos in datos]

print(datos_buenos[0][0])

for i in range(len(datos_buenos)-2):
    datos_buenos[i+1][0] = float(datos_buenos[i+1][0])
    datos_buenos[i+1][1] = float(datos_buenos[i+1][1])

datos_buenos.pop(0)
datos_buenos.pop(len(datos_buenos)-1)


Datos_energía = np.zeros(len(datos_buenos))
Datos_impulses = np.zeros(len(datos_buenos))
for i in range(len(datos_buenos)):
    Datos_energía[i] = datos_buenos[i][0]
    Datos_impulses[i] = datos_buenos[i][1]


plt.figure(figsize=(10,6))
plt.title("Fondo de radiacion")
plt.xlabel("Energía (keV)")
plt.ylabel("Impulses/#")


plt.plot(Datos_energía, Datos_impulses, marker='o', linestyle='none', color='blue', markersize=4)
plt.grid()
```

```python
with open("Muestra + fojdo", "r") as file:
    data = file.read()
    data = data.replace(',', '.')

datos = data.split('\n')
datos_buenos = [datos.split('\t') for datos in datos]

print(datos_buenos[0][0])

for i in range(len(datos_buenos)-2):
    datos_buenos[i+1][0] = float(datos_buenos[i+1][0])
    datos_buenos[i+1][1] = float(datos_buenos[i+1][1])

datos_buenos.pop(0)
datos_buenos.pop(len(datos_buenos)-1)


Datos_energía_m = np.zeros(len(datos_buenos))
Datos_impulses_m = np.zeros(len(datos_buenos))
for i in range(len(datos_buenos)):
    Datos_energía_m[i] = datos_buenos[i][0]
    Datos_impulses_m[i] = datos_buenos[i][1]

plt.figure(figsize=(10,6))
plt.title("Fondo de radiacion + Muestra")
plt.xlabel("Energía (keV)")
plt.ylabel("Impulses/#")


plt.plot(Datos_energía_m, Datos_impulses_m, marker='o', linestyle='none', color='blue', markersize=4)
plt.grid()
```

```python
plt.figure(figsize=(10,6))
plt.title("Muestra")
plt.xlabel("Energía (keV)")
plt.ylabel("Impulses/#")
plt.plot(Datos_energía_m, Datos_impulses_m-Datos_impulses, marker='o', linestyle='none', color='blue', markersize=4)
plt.grid()
```

## Ajuste Lineal

```python
datos_muestra = Datos_impulses_m - Datos_impulses

# Crear máscaras para filtrar los datos
mask = (Datos_energía_m >= 40) & (Datos_energía_m <= 150)
Datos_energía_filtrados = Datos_energía_m[mask]
datos_muestra_filtrados = datos_muestra[mask]

modelo = LinearRegression()
# Reshape para convertir a 2D array (n_samples, 1)
modelo.fit(Datos_energía_filtrados.reshape(-1, 1), datos_muestra_filtrados)
pendiente = modelo.coef_[0]
intersección = modelo.intercept_

print(f"Pendiente: {pendiente}")
print(f"Intersección: {intersección}")

```
