```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from numpy.polynomial import Polynomial
```

```python

# Abre el archivo Excel
df = pd.read_excel("/home/edgar/OneDrivePersonal/Universidad/excell.xlsx")



angulos_45=[]
angulos_30=[]
paralela_30=[]
paralela_45=[]
perpendicular_30=[]
perpendicular_45=[]

for i in range(0,16):
    angulos_45.append(df.iloc[i+1,1])
    paralela_45.append(df.iloc[i+1,5])
    perpendicular_45.append(df.iloc[i+1,6])
    if df.iloc[i+16,1] != "NaN":
        angulos_30.append(df.iloc[i+16,1])
        paralela_30.append(df.iloc[i+16,5])
        perpendicular_30.append(df.iloc[i+16,6])

coef45 = np.polyfit(angulos_45, perpendicular_45, 2)
polinomio = np.poly1d(coef45)
theta_fino = np.linspace(min(angulos_45), max(angulos_45), 200)
coef30 = np.polyfit(angulos_30, perpendicular_30, 2)
polinomio30 = np.poly1d(coef30)

```

```python

plt.figure()
fig, ax = plt.subplots()
plt.plot(theta_fino, polinomio(theta_fino),linestyle='--', label="Ajuste cuadrtico", color="red")
plt.plot(angulos_45,paralela_45,"o",label="horizontal")
plt.plot(angulos_45,perpendicular_45,"o",color="r",label="vertical")
plt.plot(57, 0.008, marker='o', markersize=10, markerfacecolor='none',markeredgecolor='black', markeredgewidth=2)
ax.annotate(r'Mnimo: $\theta = 57$', xy=(57, 0.008), xytext=(53, 0.3),arrowprops=dict(facecolor='black', edgecolor='black', shrink=0.000001),fontsize=12,bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='black'))
ax.text(76, 0, f"Ajuste cuadratico: \n"r"f($\theta$):"f"{str(polinomio[0]).split("2")[0]}"r"$\theta ^2$"f"-0.06"r"$\theta$", fontsize=10, bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='black'))
plt.title(f"GRFICA INTENSIDAD FRENTE NGULO")
plt.xlabel(r"Angulo $\alpha =45$")
plt.ylabel(f"Intensidad (I)")
plt.legend()
plt.grid()
plt.savefig("/home/edgar/GitHub/Proyectos/Proyectos_python/TEIII/graficas/grafico_45.png")


plt.figure()
fig, ax = plt.subplots()
plt.plot(theta_fino, polinomio30(theta_fino),linestyle='--', label="Ajuste cuadrtico", color="red")
plt.plot(angulos_30, paralela_30, "o", label="horizontal")
plt.plot(angulos_30, perpendicular_30, "o", color="r", label="vertical")
plt.plot(56, 0.008, marker='o', markersize=10, markerfacecolor='none',markeredgecolor='black', markeredgewidth=2)
ax.annotate(r'Mnimo: $\theta = 56$', xy=(56, 0.008), xytext=(50, 0.3),arrowprops=dict(facecolor='black', edgecolor='black', shrink=0.000001),fontsize=12,bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='black'))
ax.text(76, 0, f"Ajuste cuadratico: \n"r"f($\theta$):"f"{str(polinomio30[0]).split("3")[0]}"r"$\theta ^2$"f"{str(polinomio30[1]).split("7")[0]}"r"$\theta$", fontsize=10, bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='black'))
plt.title(f"GRFICA INTENSIDAD FRENTE NGULO")
plt.xlabel(r"Angulo $\alpha =30$")
plt.ylabel(f"Intensidad (I)")
plt.legend()
plt.grid()
```
