# Landa vs n

## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Codigo

```python
def n(theta,A=45): #n(lambda)
    theta = np.radians(theta)
    A = np.radians(A)
    
    return np.sin((A + theta)/2) / np.sin(A/2)
## Hg
theta = [91.10-110,90.50-110,90.43-110,90.5-110,90.39-110]
color = ['purple','green','orange','yellow','red']
# Verde lima
def nEdgar(theta,A=30):
    
    theta = np.radians(theta)
    return np.sin(theta)/np.sin(np.deg2rad(-13)) #n(lambda)


plt.figure(figsize=(8,6))
for i in range(len(theta)):
    plt.plot(theta[i], n(theta[i]), 'o', color=color[i]) 
    print(f'Para {color[i]}: n = {nEdgar(theta[i]):.4f}')
plt.show()

```
