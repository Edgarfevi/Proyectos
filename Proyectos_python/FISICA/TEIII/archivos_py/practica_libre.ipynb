# Landa vs n

## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Codigo

```python
def n(theta,A=45): #n(lambda)
    theta = np.radians(theta)
    A = np.radians(A)
    
    return np.sin((A + theta)/2) / np.sin(A/2)
## Hg 83.6
theta = [83.6-55.6,83.6-56,83.6-56.15]
color = ['purple','green','orange']


plt.figure(figsize=(8,6))
for i in range(len(theta)):
    plt.plot(theta[i], n(theta[i]), 'o', color=color[i])
    plt.xlabel('theta (Â°)')
    plt.ylabel('n(lambda)')
    plt.title('Refractive index vs angle of incidence')
    plt.grid()
    print(f'Para {color[i]}: n = {n(theta[i]):.4f}')
plt.show()

```
