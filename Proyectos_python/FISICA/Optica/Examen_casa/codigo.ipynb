## Calculo de alpha 2 con ley de snell

```python
def Ley_snell(n1,n2,alpha1):
    alpha2 = asin((n1/n2)*sin(radians(alpha1)))
    return degrees(alpha2)
```

```python
variable_continuacion=str(input("quieres continuar? s/n "))
i=0
n1=0
n2=0
alpha1=0
alpha2=0

try:
    while variable_continuacion=="s":
        if i == 0:
            n1=float(input("ingresa el indice de refraccion del primer medio: "))
            n2=float(input("ingresa el indice de refraccion del segundo medio: "))
            alpha1=float(input("ingresa el angulo de incidencia en grados: "))
            alpha2=Ley_snell(n1,n2,alpha1)
            print(f"el angulo de refraccion es: {alpha2} grados")
            variable_continuacion=str(input("quieres continuar? s/n "))
            i+=1
            
        else:
            n1=n2
            n2=float(input("ingresa el indice de refraccion del siguiente medio: "))
            alpha1=float(input("ingresa el angulo de incidencia en grados: "))
            alpha2=Ley_snell(n1,n2,alpha1)
            print(f"el angulo de refraccion es: {alpha2} grados")
            variable_continuacion=str(input("quieres continuar? s/n "))
except ValueError:
    print("reflexi√≥n total interna ")
```
