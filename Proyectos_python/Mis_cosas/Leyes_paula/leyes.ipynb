```python
import re

def parsear_articulos(texto):
    """
    Busca patrones 'Artículo N. Texto...' y devuelve un diccionario.
    Clave: Número del artículo (string).
    Valor: Texto completo del artículo.
    """
    articulos = {}
    
    # EXPLICACIÓN DEL REGEX (PATRÓN):
    # ^\s*Artículo\s+(\d+)  -> Busca líneas que empiecen por 'Artículo' seguido de un número (capturamos el número).
    # \.                    -> Un punto literal después del número.
    # .*?                   -> Coge todo el texto que siga (contenido del artículo).
    # (?=\n\s*Artículo|\Z)  -> Para (Lookahead) cuando encuentres el siguiente 'Artículo' o el final del archivo.
    
    patron = re.compile(r'(^\s*Artículo\s+(\d+)\..*?)(?=\n\s*Artículo\s+\d+\.|\Z)', re.MULTILINE | re.DOTALL)
    
    for match in patron.finditer(texto):
        contenido_completo = match.group(1).strip()
        numero_articulo = match.group(2) # El número '10', '25', etc.
        
        articulos[numero_articulo] = contenido_completo
        
    return articulos

def consolidar_ley(ruta_original, ruta_reforma, ruta_salida):
    # 1. Leer los archivos
    with open(ruta_original, 'r', encoding='utf-8') as f:
        txt_original = f.read()
    
    with open(ruta_reforma, 'r', encoding='utf-8') as f:
        txt_reforma = f.read()

    # 2. Convertir texto a diccionarios
    dict_original = parsear_articulos(txt_original)
    dict_reforma = parsear_articulos(txt_reforma)
    
    print(f"Artículos encontrados en original: {len(dict_original)}")
    print(f"Artículos encontrados en reforma: {len(dict_reforma)}")

    # 3. Sustitución (El corazón del programa)
    cambios_realizados = 0
    for num, nuevo_texto in dict_reforma.items():
        if num in dict_original:
            # Aquí ocurre la magia: reemplazamos el viejo por el nuevo
            dict_original[num] = nuevo_texto
            print(f"Actualizado: Artículo {num}")
            cambios_realizados += 1
        else:
            # Opcional: Si el artículo es nuevo (no existía antes), lo añadimos
            # Nota: Esto lo añade al final del diccionario, ordenarlo después es otro reto.
            dict_original[num] = nuevo_texto 
            print(f"Añadido nuevo (o no encontrado en original): Artículo {num}")

    # 4. Reconstruir el texto final
    # Para que salga ordenado numéricamente y no caótico:
    texto_final = ""
    
    # Ordenamos las claves numéricamente (1, 2, 3... 10) en vez de alfabéticamente (1, 10, 2...)
    claves_ordenadas = sorted(dict_original.keys(), key=lambda x: int(x) if x.isdigit() else 9999)
    
    for key in claves_ordenadas:
        texto_final += dict_original[key] + "\n\n"

    # 5. Guardar
    with open(ruta_salida, 'w', encoding='utf-8') as f:
        f.write(texto_final)
    
    print(f"--- Proceso terminado. {cambios_realizados} artículos actualizados. ---")

# --- EJECUCIÓN ---
# Asegúrate de poner los nombres de tus archivos reales aquí
try:
    consolidar_ley("texto_extraido.txt", "texto_extraido_202.txt", "ley_consolidada.txt")
except FileNotFoundError:
    print("Error: No se encuentran los archivos .txt. Revisa los nombres.")
```

```python
from fpdf import FPDF

def crear_pdf_desde_txt(archivo_entrada, archivo_salida):
    # 1. Configuración básica del PDF
    pdf = FPDF()
    pdf.add_page()
    
    # Elegimos fuente. Arial es estándar.
    # Tamaño 12 es bueno para lectura legal.
    pdf.set_font("Arial", size=11)

    # 2. Leer el archivo TXT
    try:
        with open(archivo_entrada, "r", encoding="utf-8") as f:
            texto = f.read()
    except FileNotFoundError:
        print("¡Error! No encuentro el archivo .txt")
        return

    # 3. TRUCO PARA ACENTOS Y 'Ñ' (Importante)
    # FPDF por defecto no soporta UTF-8 directo con las fuentes estándar.
    # Tenemos que codificarlo a 'latin-1' y reemplazar caracteres incompatibles para que no falle.
    texto_para_pdf = texto.encode('latin-1', 'replace').decode('latin-1')

    # 4. Escribir el texto
    # multi_cell(ancho, alto_linea, texto)
    # w=0 significa "usa todo el ancho de la página"
    # h=5 es la altura entre líneas (interlineado)
    # align='J' justifica el texto (queda más profesional, como el BOE real)
    pdf.multi_cell(w=0, h=6, txt=texto_para_pdf, align='J')

    # 5. Guardar
    pdf.output(archivo_salida)
    print(f"¡Éxito! PDF guardado como: {archivo_salida}")

# --- Ejecución ---
crear_pdf_desde_txt("ley_consolidada.txt", "ley_final_boe.pdf")
```

```python
import pdfplumber
import re

def analizar_reformas(archivo_pdf):
    print(f"--- Analizando: {archivo_pdf} ---\n")
    
    texto_completo = ""

    # 1. PASO PREVIO: Convertir PDF a Texto (respetando flujo de lectura)
    try:
        with pdfplumber.open(archivo_pdf) as pdf:
            for pagina in pdf.pages:
                # Extract_text de pdfplumber une las columnas inteligentemente
                texto_pagina = pagina.extract_text()
                if texto_pagina:
                    texto_completo += texto_pagina + "\n"
    except FileNotFoundError:
        print("Error: No encuentro el archivo PDF.")
        return

    # Limpiamos saltos de línea excesivos para facilitar la búsqueda
    # Convertimos todo a una sola línea larga o párrafos limpios
    texto_limpio = re.sub(r'\s+', ' ', texto_completo)

    # 2. DEFINIR LOS PATRONES DE BÚSQUEDA (REGEX)
    
    # Explicación del Regex para MODIFICACIONES:
    # (?i)          -> Ignorar mayúsculas/minúsculas
    # Se modifica   -> La frase clave
    # .*?           -> Cualquier texto en medio (ej: "el", "el contenido del")
    # art[íi]culo   -> Busca "articulo" con o sin tilde
    # \s+           -> Espacio
    # (\d+\s*\w*)   -> CAPTURA el número (y letras como 'bis' o 'ter')
    patron_modif = r"(?i)se modifica.*?art[íi]culo\s+(\d+\s*\w*)"
    
    # Explicación del Regex para INTRODUCCIONES / AÑADIDOS:
    # Busca variaciones: "Se introduce", "Se añade", "Se incorpora"
    patron_intro = r"(?i)(?:se introduce|se a[ñn]ade|se incorpora).*?art[íi]culo\s+(\d+\s*\w*)"

    # 3. BUSCAR Y MOSTRAR RESULTADOS

    # Buscamos todas las coincidencias
    modificaciones = re.findall(patron_modif, texto_limpio)
    introducciones = re.findall(patron_intro, texto_limpio)

    print(f"RESUMEN DETECTADO:")
    print(f"==================")
    
    # Imprimir Modificaciones
    if modificaciones:
        print(f"\n[!] Artículos que se MODIFICAN ({len(modificaciones)}):")
        for art in modificaciones:
            print(f" -> Artículo {art}")
    else:
        print("\nNo se detectaron frases de 'Se modifica el artículo...'.")

    # Imprimir Introducciones
    if introducciones:
        print(f"\n[+] Artículos que se INTRODUCEN/AÑADEN ({len(introducciones)}):")
        for art in introducciones:
            print(f" -> Artículo {art}")
    else:
        print("\nNo se detectaron frases de 'Se introduce/añade el artículo...'.")

# --- EJECUCIÓN ---
# Cambia 'documento_2022.pdf' por el nombre de tu archivo real
analizar_reformas("BOE/BOE-A-2022-14580-consolidado.pdf")
```

```python

def analizar_detalle_boe(archivo_pdf):
    print(f"--- Analizando documento: {archivo_pdf} ---\n")
    
    texto_completo = ""

    # 1. Extracción del texto del PDF
    try:
        with pdfplumber.open(archivo_pdf) as pdf:
            for pagina in pdf.pages:
                texto_pagina = pagina.extract_text()
                if texto_pagina:
                    texto_completo += texto_pagina + " " # Añadimos espacio para evitar pegados
    except FileNotFoundError:
        print("Error: No encuentro el archivo PDF.")
        return

    # Limpieza: Convertimos saltos de línea y espacios múltiples en un solo espacio
    texto_limpio = re.sub(r'\s+', ' ', texto_completo)

    # 2. EL SUPER-REGEX
    # Explicación de los grupos de captura ():
    # Grupo 1 (Acción): Detecta "Se modifica", "Se añade", etc.
    # Grupo 2 (Intermedio): Captura TODO lo que hay entre la acción y la palabra "artículo".
    # Grupo 3 (Artículo): Captura el número del artículo.
    
    patron = r"(?i)(se modifica|se introduce|se a[ñn]ade|se incorpora|se da nueva redacción a)(.*?)art[íi]culo\s+(\d+\s*\w*)"

    coincidencias = re.finditer(patron, texto_limpio)
    
    # Listas para guardar los resultados clasificados
    cambios_art_completos = []
    cambios_apartados = []

    print("Procesando hallazgos...\n")

    for match in coincidencias:
        accion = match.group(1).strip()       # Ej: "Se modifica"
        texto_intermedio = match.group(2)     # Ej: " el apartado 3 del "
        num_articulo = match.group(3)         # Ej: "20"

        # 3. LÓGICA DE CLASIFICACIÓN
        # Analizamos el texto intermedio buscando palabras clave de "trozos"
        
        palabras_clave_parcial = ["apartado", "párrafo", "letra", "punto", "número"]
        
        # Comprobamos si alguna palabra clave está en el texto intermedio
        es_parcial = any(palabra in texto_intermedio.lower() for palabra in palabras_clave_parcial)

        if es_parcial:
            # Intentamos limpiar el texto intermedio para que se lea bien
            # Ej: de " el apartado 2 del " pasamos a "apartado 2"
            detalle = texto_intermedio.strip().replace(" el ", "").replace(" del ", "").replace(" al ", "")
            cambios_apartados.append(f"Art. {num_articulo} -> Se toca: '{detalle}' ({accion})")
        else:
            cambios_art_completos.append(f"Art. {num_articulo} -> Completo ({accion})")

    # 4. IMPRIMIR RESULTADOS
    print(f"=== RESUMEN DE CAMBIOS ===")
    
    if cambios_art_completos:
        print(f"\n[O] ARTÍCULOS COMPLETOS O PRINCIPALES ({len(cambios_art_completos)}):")
        for item in cambios_art_completos:
            print(f"  - {item}")
    
    if cambios_apartados:
        print(f"\n[>] APARTADOS O SECCIONES ESPECÍFICAS ({len(cambios_apartados)}):")
        for item in cambios_apartados:
            print(f"  - {item}")
            
    if not cambios_art_completos and not cambios_apartados:
        print("No se han encontrado coincidencias con los patrones estándar.")

# --- EJECUCIÓN ---
analizar_detalle_boe("BOE/BOE-A-2022-14580-consolidado.pdf")
```

```python
import re
from fpdf import FPDF

# --- CONFIGURACIÓN ---
RANGO_MIN = 100
RANGO_MAX = 504
ARCHIVO_ORIGINAL = "ley_original.txt"
ARCHIVO_REFORMA = "ley_reforma.txt"
ARCHIVO_SALIDA = "Ley_Consolidada_Final.pdf"

class ConsolidadorBOE:
    def __init__(self):
        self.articulos = {} 

    def leer_original(self, ruta):
        print(f"--- Leyendo original: {ruta} ---")
        with open(ruta, 'r', encoding='utf-8') as f:
            texto = f.read()

        # Regex para capturar artículos originales
        patron = re.compile(r'(^\s*Art[íi]culo\s+(\d+)\.?.*?)((?=\n\s*Art[íi]culo\s+\d+)|\Z)', re.MULTILINE | re.DOTALL | re.IGNORECASE)

        count = 0
        for match in patron.finditer(texto):
            contenido = match.group(1).strip()
            numero = int(match.group(2))
            self.articulos[numero] = contenido
            count += 1
        print(f" -> Cargados {count} artículos del original.")

    def aplicar_reformas(self, ruta_reforma):
        print(f"\n--- Procesando reformas: {ruta_reforma} ---")
        with open(ruta_reforma, 'r', encoding='utf-8') as f:
            texto_reforma = f.read()

        # Regex para detectar bloques en el archivo de reforma
        # Detecta: ARTÍCULO 100 [APARTADO 2]
        patron_bloque = re.compile(r'(^\s*Art[íi]culo\s+(\d+)(?:\s+Apartado\s+(\d+))?\.?)(.*?)((?=\n\s*Art[íi]culo)|\Z)', re.MULTILINE | re.DOTALL | re.IGNORECASE)

        for match in patron_bloque.finditer(texto_reforma):
            cabecera = match.group(1).strip() # "ARTÍCULO 100"
            num_art = int(match.group(2))     # 100
            num_apartado = match.group(3)     # 2 (o None si es todo el art)
            contenido_nuevo = match.group(4).strip() # El texto de abajo

            # 1. FILTRO DE RANGO
            if not (RANGO_MIN <= num_art <= RANGO_MAX):
                continue

            # Detectamos si la orden es SUPRIMIR
            es_supresion = "SUPRIMIDO" in contenido_nuevo.upper() and len(contenido_nuevo) < 20

            # --- CASO A: ARTÍCULO COMPLETO ---
            if num_apartado is None:
                if es_supresion:
                    # Borrar artículo entero
                    if num_art in self.articulos:
                        del self.articulos[num_art]
                        print(f"[-] Art. {num_art}: ELIMINADO completamente.")
                    else:
                        print(f"[!] Art. {num_art}: Se intentó borrar pero no existía.")
                else:
                    # Modificar/Añadir artículo entero
                    accion = "Modificado" if num_art in self.articulos else "Añadido nuevo"
                    self.articulos[num_art] = f"Artículo {num_art}. {contenido_nuevo}"
                    print(f"[+] Art. {num_art}: {accion} (Completo).")

            # --- CASO B: APARTADO ESPECÍFICO ---
            else:
                if num_art not in self.articulos:
                    print(f"[!] Error: No puedo tocar el apartado {num_apartado} del Art {num_art} porque el artículo padre no existe.")
                    continue
                
                texto_actual = self.articulos[num_art]
                
                # Regex para buscar el apartado específico dentro del texto actual
                # Busca: Salto de línea + Número + punto/guión/paréntesis
                regex_apartado = fr"(\n\s*{num_apartado}[\.\-\)].*?)(?=\n\s*\d+[\.\-\)]|\Z)"
                
                match_apartado = re.search(regex_apartado, texto_actual, re.DOTALL)

                if match_apartado:
                    if es_supresion:
                        # OPCIÓN 1: Dejar rastro legal (Recomendado)
                        reemplazo = f"\n{num_apartado}. (Suprimido)"
                        # OPCIÓN 2: Borrarlo del mapa (Descomenta la siguiente línea si prefieres que desaparezca)
                        # reemplazo = "" 
                        
                        nuevo_texto = re.sub(regex_apartado, reemplazo, texto_actual, count=1, flags=re.DOTALL)
                        self.articulos[num_art] = nuevo_texto
                        print(f"[-] Art. {num_art}: Apartado {num_apartado} SUPRIMIDO.")
                    
                    else:
                        # Modificar contenido del apartado
                        texto_limpio_a_pegar = f"{num_apartado}. {contenido_nuevo}"
                        nuevo_texto = re.sub(regex_apartado, f"\n{texto_limpio_a_pegar}", texto_actual, count=1, flags=re.DOTALL)
                        self.articulos[num_art] = nuevo_texto
                        print(f"[~] Art. {num_art}: Apartado {num_apartado} MODIFICADO.")
                else:
                    print(f"[!] Aviso: No encontré el apartado {num_apartado} en el Art {num_art}. No se hizo nada.")

    def generar_pdf(self, ruta_salida):
        print(f"\n--- Generando PDF: {ruta_salida} ---")
        pdf = FPDF()
        pdf.set_auto_page_break(auto=True, margin=15)
        pdf.add_page()
        pdf.set_font("Arial", size=11)
        
        pdf.set_font("Arial", "B", 16)
        pdf.cell(0, 10, "TEXTO CONSOLIDADO", ln=True, align='C')
        pdf.set_font("Arial", "I", 10)
        pdf.cell(0, 10, f"Artículos del {RANGO_MIN} al {RANGO_MAX}", ln=True, align='C')
        pdf.ln(10)

        # Ordenar claves numéricamente
        for num in sorted(self.articulos.keys()):
            if RANGO_MIN <= num <= RANGO_MAX:
                texto = self.articulos[num]
                
                # Gestión de codificación para PDF
                texto_safe = texto.encode('latin-1', 'replace').decode('latin-1')
                
                # Intentamos separar título y cuerpo para negritas
                lineas = texto_safe.split('\n', 1)
                
                pdf.set_font("Arial", "B", 11)
                # Escribimos la primera línea (ej: Artículo 100.)
                pdf.multi_cell(0, 6, lineas[0], align='J') 
                
                if len(lineas) > 1:
                    pdf.set_font("Arial", "", 11)
                    pdf.multi_cell(0, 6, lineas[1], align='J')
                
                pdf.ln(5) # Espacio entre artículos

        pdf.output(ruta_salida)
        print("¡Proceso Terminado con éxito!")

# --- EJECUCIÓN ---
gestor = ConsolidadorBOE()

try:
    gestor.leer_original(ARCHIVO_ORIGINAL)
    gestor.aplicar_reformas(ARCHIVO_REFORMA)
    gestor.generar_pdf(ARCHIVO_SALIDA)
except FileNotFoundError as e:
    print(f"\n[ERROR CRÍTICO]: Faltan archivos. {e}")
```

```python
import pdfplumber
import re
from fpdf import FPDF

# --- CONFIGURACIÓN ---
PDF_ORIGINAL = "BOE/BOE-A-2020-4859-consolidado.pdf"
PDF_REFORMA = "BOE/BOE-A-2022-14580-consolidado.pdf"
PDF_SALIDA = "Ley_Concursal_2022_Final.pdf"

class FusionadorLegalPro:
    def __init__(self):
        self.articulos = {}  # { "20": "Texto completo..." }
        self.ultimo_art_visto = None # Memoria para saber a qué artículo pertenece un apartado suelto

    def limpiar_basura_boe(self, texto):
        """Elimina cabeceras, pies de página y une líneas cortadas."""
        lineas = texto.split('\n')
        lineas_limpias = []
        for linea in lineas:
            # Filtros de ruido habitual en BOE
            if any(x in linea for x in ["BOLETÍN OFICIAL", "LEGISLACIÓN CONSOLIDADA", "Página ", "Referencia: BOE"]):
                continue
            lineas_limpias.append(linea)
        
        texto_limpio = "\n".join(lineas_limpias)
        # Unir palabras cortadas por guiones al final de línea (ej: ins- \n truccción)
        texto_limpio = re.sub(r'(\w+)-\n(\w+)', r'\1\2', texto_limpio)
        return texto_limpio

    def procesar_original(self):
        """Lee el documento base (2020)"""
        print(f"--- Leyendo Base: {PDF_ORIGINAL} ---")
        texto = self.extraer_texto_pdf(PDF_ORIGINAL)
        
        # Regex Estándar para artículos originales (sin comillas normalmente)
        patron = re.compile(r'(?:^|\n)\s*Art[íi]culo\s+(\d+)\.?(.*?)((?=\n\s*Art[íi]culo)|\Z)', re.DOTALL | re.IGNORECASE)

        for match in patron.finditer(texto):
            num = match.group(1)
            contenido = f"Artículo {num}.{match.group(2)}"
            self.articulos[num] = contenido.strip()
        
        print(f"   -> Cargados {len(self.articulos)} artículos base.")

    def procesar_reforma(self):
        """Lee la reforma y busca tanto artículos enteros como apartados sueltos entre comillas"""
        print(f"--- Leyendo Reforma: {PDF_REFORMA} ---")
        texto = self.extraer_texto_pdf(PDF_REFORMA)
        
        # 1. DIVIDIMOS EL TEXTO EN BLOQUES PARA ANALIZAR CONTEXTO
        # (Esto ayuda a saber que si vemos un "2.", pertenece al artículo mencionado antes)
        bloques = re.split(r'(Se modifica.*?Art[íi]culo\s+\d+)', texto, flags=re.IGNORECASE)
        
        art_actual_contexto = None

        # REGEX MEJORADO: Acepta <<, «, ", o nada al principio
        regex_articulo_nuevo = r'(?:«|<<|“|")\s*Art[íi]culo\s+(\d+)\.?(.*?)(?:»|>>|”|")'
        regex_apartado_nuevo = r'(?:«|<<|“|")\s*(\d+)[\.\-](.*?)(?:»|>>|”|")'
        
        # Búsqueda general de menciones a artículos para el contexto (ej: "Se modifica el artículo 20")
        regex_contexto = r'Art[íi]culo\s+(\d+)'

        for bloque in bloques:
            # A. Actualizar contexto (¿De qué artículo estamos hablando en este párrafo?)
            match_contexto = re.search(regex_contexto, bloque, re.IGNORECASE)
            if match_contexto:
                art_actual_contexto = match_contexto.group(1)

            # B. Buscar ARTÍCULO COMPLETO NUEVO (con comillas)
            # Ej: <<Artículo 20. El deudor...>>
            for match in re.finditer(regex_articulo_nuevo, bloque, re.DOTALL | re.IGNORECASE):
                num = match.group(1)
                contenido_raw = match.group(2)
                
                # Limpiamos las comillas de cierre si se colaron
                contenido_limpio = contenido_raw.replace('»', '').replace('>>', '')
                self.articulos[num] = f"Artículo {num}. {contenido_limpio}".strip()
                print(f"   [Cambio Total] Artículo {num} actualizado.")

            # C. Buscar APARTADO SUELTO (con comillas)
            # Ej: <<4. La solicitud se presentará...>>
            # Solo funciona si sabemos en qué artículo estamos (art_actual_contexto)
            if art_actual_contexto and art_actual_contexto in self.articulos:
                for match in re.finditer(regex_apartado_nuevo, bloque, re.DOTALL):
                    num_apartado = match.group(1)
                    contenido_apartado = match.group(2).replace('»', '').replace('>>', '')
                    
                    texto_viejo = self.articulos[art_actual_contexto]
                    
                    # Intentamos inyectar el apartado nuevo
                    # Buscamos dónde está ese número de apartado en el texto viejo
                    # Regex busca: \n3. o \n3- 
                    regex_hueco = fr"(\n\s*{num_apartado}[\.\-].*?)(?=\n\s*\d+[\.\-]|\Z)"
                    
                    if re.search(regex_hueco, texto_viejo):
                        nuevo_bloque = f"\n{num_apartado}. {contenido_apartado}"
                        texto_actualizado = re.sub(regex_hueco, nuevo_bloque, texto_viejo, count=1, flags=re.DOTALL)
                        self.articulos[art_actual_contexto] = texto_actualizado
                        print(f"   [Cambio Parcial] Art {art_actual_contexto}, Apartado {num_apartado}.")

    def extraer_texto_pdf(self, ruta):
        full_text = ""
        with pdfplumber.open(ruta) as pdf:
            for page in pdf.pages:
                extracted = page.extract_text()
                if extracted: full_text += extracted + "\n"
        return self.limpiar_basura_boe(full_text)

    def generar_pdf(self):
        print(f"--- Generando PDF Final... ---")
        pdf = FPDF()
        pdf.set_auto_page_break(auto=True, margin=15)
        pdf.add_page()
        
        pdf.set_font("Arial", "B", 16)
        pdf.cell(0, 10, "LEY CONCURSAL CONSOLIDADA 2022", ln=True, align='C')
        pdf.ln(10)

        # Ordenar numéricamente las claves (string "10" -> int 10)
        claves = sorted(self.articulos.keys(), key=lambda x: int(x) if x.isdigit() else 9999)

        for num in claves:
            if not num.isdigit(): continue # Saltamos cosas raras
            
            texto = self.articulos[num]
            
            # Gestión de codificación
            texto_safe = texto.encode('latin-1', 'replace').decode('latin-1')
            
            # Título en negrita
            partes = texto_safe.split('.', 1)
            titulo = partes[0] + "."
            cuerpo = partes[1] if len(partes) > 1 else ""

            pdf.set_font("Arial", "B", 11)
            pdf.multi_cell(0, 6, titulo, align='J')
            
            pdf.set_font("Arial", "", 10)
            pdf.multi_cell(0, 5, cuerpo, align='J')
            pdf.ln(4)

        pdf.output(PDF_SALIDA)
        print(f"¡Éxito! Archivo creado: {PDF_SALIDA}")

if __name__ == "__main__":
    bot = FusionadorLegalPro()
    bot.procesar_original()
    bot.procesar_reforma()
    bot.generar_pdf()
```

```python
import pdfplumber
import re
from fpdf import FPDF

# --- TUS ARCHIVOS ---
PDF_ORIGINAL = "BOE/BOE-A-2020-4859-consolidado.pdf"
PDF_REFORMA = "BOE/BOE-A-2022-14580-consolidado.pdf"
PDF_SALIDA = "Ley_Concursal_Consolidada_Limpia.pdf"

class ConsolidadorBOE_Pro:
    def __init__(self):
        self.articulos = {} 

    def limpiar_texto_boe(self, texto):
        """Elimina cabeceras y pies de página recurrentes."""
        lineas = texto.split('\n')
        lineas_limpias = []
        for linea in lineas:
            # Filtro estricto de basura de encabezado
            if any(x in linea for x in ["BOLETÍN OFICIAL", "LEGISLACIÓN CONSOLIDADA", "Página ", "Referencia: BOE"]):
                continue
            lineas_limpias.append(linea)
        return "\n".join(lineas_limpias)

    def cargar_original(self):
        print(f"--- 1. Leyendo Original: {PDF_ORIGINAL} ---")
        texto_completo = ""
        with pdfplumber.open(PDF_ORIGINAL) as pdf:
            for page in pdf.pages:
                texto_completo += page.extract_text() + "\n"
        
        texto_limpio = self.limpiar_texto_boe(texto_completo)
        
        # Regex para capturar artículos del 2020 (Artículos sin comillas)
        patron = re.compile(r'(?:^|\n)\s*Art[íi]culo\s+(\d+)\.?(.*?)((?=\n\s*Art[íi]culo)|\Z)', re.DOTALL | re.IGNORECASE)
        
        for match in patron.finditer(texto_limpio):
            num = match.group(1)
            contenido = f"Artículo {num}.{match.group(2)}"
            self.articulos[num] = contenido.strip()
        print(f"   -> Artículos base cargados: {len(self.articulos)}")

    def aplicar_reforma(self):
        print(f"--- 2. Procesando Reforma (Modo Quirúrgico): {PDF_REFORMA} ---")
        texto_reforma = ""
        with pdfplumber.open(PDF_REFORMA) as pdf:
            for page in pdf.pages:
                texto_reforma += page.extract_text() + "\n"
        
        texto_reforma = self.limpiar_texto_boe(texto_reforma)

        # PASO 1: Dividir la reforma por instrucciones de modificación
        # Buscamos frases tipo "Se modifica el artículo X" o "Se añade el artículo Y"
        # Esto nos da el "Contexto" para no perdernos.
        regex_instruccion = r'(Se (?:modifica|añade|introduce|incorpora|da nueva redacción).*?Art[íi]culo\s+(\d+))'
        
        bloques = re.split(regex_instruccion, texto_reforma, flags=re.IGNORECASE)
        
        # Iteramos: Bloque 0 es texto, Bloque 1 es la instrucción, Bloque 2 es el numero art, Bloque 3 es el contenido...
        # re.split con grupos de captura devuelve: [texto_antes, instruccion, numero, texto_despues, instruccion, numero...]
        
        i = 1
        while i < len(bloques) - 1:
            instruccion = bloques[i]      # Ej: "Se modifica el artículo 20"
            num_art = bloques[i+1]        # Ej: "20"
            contenido_bloque = bloques[i+2] # El texto que viene después
            
            # Limpiamos el contenido del bloque hasta la siguiente instrucción para analizarlo
            # (Aunque el split ya lo hace, aseguramos limpieza)
            
            # --- ANÁLISIS DEL BLOQUE ---
            
            # A. ¿Es una SUPRESIÓN?
            if "suprime" in instruccion.lower() or "deroga" in instruccion.lower() or "deja sin contenido" in instruccion.lower():
                # Verificamos si es apartado o artículo entero
                if "apartado" in instruccion.lower() or "número" in instruccion.lower() or "letra" in instruccion.lower():
                    # Es parcial, necesitamos saber qué apartado es.
                    # Buscamos "apartado X" en la instrucción
                    match_ap = re.search(r'(?:apartado|número|punto)\s+(\d+|[a-z]\))', instruccion, re.IGNORECASE)
                    if match_ap and num_art in self.articulos:
                        ap_borrar = match_ap.group(1)
                        print(f"   [-] Art. {num_art}: SUPRIMIDO Apartado {ap_borrar}")
                        # Lo marcamos en el texto como derogado
                        self.borrar_apartado_en_texto(num_art, ap_borrar)
                else:
                    # Es total
                    if num_art in self.articulos:
                        del self.articulos[num_art]
                        print(f"   [X] Art. {num_art}: ELIMINADO COMPLETO")
            
            # B. ¿Hay contenido NUEVO entre comillas? (Modificación o Adición)
            else:
                # Buscamos SOLO texto entre comillas BOE («...» o “...”)
                # Esto es lo que filtra tu error del "Art 26 y el Preámbulo"
                regex_comillas = r'(?:«|<<|“|")\s*(.*?)(?:»|>>|”|")'
                match_contenido = re.search(regex_comillas, contenido_bloque, re.DOTALL)
                
                if match_contenido:
                    texto_nuevo = match_contenido.group(1).strip()
                    
                    # ¿Es un Artículo Completo o un Apartado?
                    # Si el texto nuevo empieza por "Artículo X." -> Completo
                    if re.match(r'^Art[íi]culo\s+\d+', texto_nuevo, re.IGNORECASE):
                        self.articulos[num_art] = texto_nuevo
                        print(f"   [O] Art. {num_art}: Reemplazo COMPLETO")
                    
                    # Si empieza por número o letra (ej: "4." o "a)") -> Apartado
                    elif re.match(r'^(\d+|[a-z]\))[\.\-\)]', texto_nuevo):
                        match_num = re.match(r'^(\d+|[a-z]\))', texto_nuevo)
                        num_ap = match_num.group(1)
                        if num_art in self.articulos:
                            self.inyectar_apartado(num_art, num_ap, texto_nuevo)
                            print(f"   [>] Art. {num_art}: Modificado Apartado {num_ap}")
                        else:
                            # Si el artículo no existía y es un apartado, lo creamos (raro pero posible)
                            self.articulos[num_art] = f"Artículo {num_art}.\n{texto_nuevo}"
                            print(f"   [+] Art. {num_art}: Creado nuevo con Apartado {num_ap}")
                    else:
                        # A veces es texto suelto que sustituye un párrafo sin número
                        # Asumimos reemplazo completo si hay duda o es adición
                        if "se añade" in instruccion.lower() and num_art not in self.articulos:
                             self.articulos[num_art] = f"Artículo {num_art}. {texto_nuevo}"
                             print(f"   [+] Art. {num_art}: AÑADIDO NUEVO")

            i += 3 # Saltamos al siguiente grupo

    def borrar_apartado_en_texto(self, num_art, num_apartado):
        texto = self.articulos[num_art]
        # Regex busca: \n4. Texto... hasta el siguiente \n5.
        regex = fr"(\n\s*{re.escape(num_apartado)}[\.\-\)].*?)(?=\n\s*(?:\d+|[a-z])[\.\-\)]|\Z)"
        self.articulos[num_art] = re.sub(regex, f"\n{num_apartado}. (Suprimido)", texto, flags=re.DOTALL)

    def inyectar_apartado(self, num_art, num_apartado, texto_nuevo):
        texto = self.articulos[num_art]
        # Regex para encontrar dónde está ese apartado actualmente
        regex = fr"(\n\s*{re.escape(num_apartado)}[\.\-\)].*?)(?=\n\s*(?:\d+|[a-z])[\.\-\)]|\Z)"
        
        if re.search(regex, texto):
            # Reemplazar
            self.articulos[num_art] = re.sub(regex, f"\n{texto_nuevo}", texto, count=1, flags=re.DOTALL)
        else:
            # Si no existe, lo añadimos al final (mejor que perderlo)
            self.articulos[num_art] += f"\n{texto_nuevo}"

def generar_pdf(self):
        print(f"--- 3. Generando PDF Final: {PDF_SALIDA} ---")
        pdf = FPDF()
        pdf.set_auto_page_break(auto=True, margin=15)
        pdf.add_page()
        pdf.set_font("Arial", "B", 16)
        pdf.cell(0, 10, "LEY CONCURSAL CONSOLIDADA (Oficial)", ln=True, align='C')
        pdf.ln(10)

        # --- FUNCIÓN MAESTRA DE ORDENACIÓN ---
        def obtener_valor_orden(clave_articulo):
            # 1. Extraemos el número base (Ej: de "58 bis" saca 58)
            match = re.match(r"(\d+)", clave_articulo)
            if not match: 
                return 999999 # Si es una Disposición o algo raro, al final del todo

            numero = int(match.group(1))
            decimal = 0.0

            # 2. Asignamos valor decimal según el sufijo latino
            # Esto hace que 58 bis (58.01) vaya después de 58 (58.00) pero antes de 59 (59.00)
            clave_lower = clave_articulo.lower()
            
            if "bis" in clave_lower:       decimal = 0.01
            elif "ter" in clave_lower:     decimal = 0.02
            elif "quater" in clave_lower:  decimal = 0.03
            elif "quinquies" in clave_lower: decimal = 0.04
            elif "sexies" in clave_lower:  decimal = 0.05
            elif "septies" in clave_lower: decimal = 0.06
            elif "octies" in clave_lower:  decimal = 0.07
            elif "nonies" in clave_lower:  decimal = 0.08
            elif "decies" in clave_lower:  decimal = 0.09
            elif "undecies" in clave_lower: decimal = 0.10
            elif "duodecies" in clave_lower: decimal = 0.11
            
            return numero + decimal

        # Ordenamos las claves usando nuestra función maestra
        claves_ordenadas = sorted(self.articulos.keys(), key=obtener_valor_orden)

        for num in claves_ordenadas:
            texto = self.articulos[num]
            try:
                texto_safe = texto.encode('latin-1', 'replace').decode('latin-1')
            except:
                continue 

            # Formato bonito
            lineas = texto_safe.split('.', 1)
            
            # Título (Ej: Artículo 58 bis.)
            pdf.set_font("Arial", "B", 11)
            titulo = lineas[0] + "."
            pdf.multi_cell(0, 6, titulo, align='J')
            
            # Cuerpo
            if len(lineas) > 1:
                pdf.set_font("Arial", "", 10)
                pdf.multi_cell(0, 5, lineas[1].strip(), align='J')
            pdf.ln(4)

        pdf.output(PDF_SALIDA)
        print("¡Proceso finalizado!")
        # Ordenar claves correctamente (incluyendo bis/ter)
        def ordenar(k):
            m = re.match(r"(\d+)", k)
            if m:
                val = int(m.group(1))
                if "bis" in k: val += 0.1
                if "ter" in k: val += 0.2
                return val
            return 99999

        for num in sorted(self.articulos.keys(), key=ordenar):
            texto = self.articulos[num]
            try:
                texto_safe = texto.encode('latin-1', 'replace').decode('latin-1')
            except:
                continue 

            # Formato bonito
            lineas = texto_safe.split('.', 1)
            pdf.set_font("Arial", "B", 11)
            pdf.multi_cell(0, 6, lineas[0] + ".", align='J')
            
            if len(lineas) > 1:
                pdf.set_font("Arial", "", 10)
                pdf.multi_cell(0, 5, lineas[1].strip(), align='J')
            pdf.ln(4)

        pdf.output(PDF_SALIDA)
        print("¡Proceso finalizado!")

if __name__ == "__main__":
    app = ConsolidadorBOE_Pro()
    app.cargar_original()
    app.aplicar_reforma()
    app.generar_pdf()
```

```python
import pdfplumber
import re
from fpdf import FPDF

# --- CONFIGURACIÓN DE ARCHIVOS ---
PDF_ORIGINAL = "BOE/BOE-A-2020-4859-consolidado.pdf"
PDF_REFORMA = "BOE/BOE-A-2022-14580-consolidado.pdf"
PDF_SALIDA = "Ley_Concursal_Consolidada_Final.pdf"

class ConsolidadorLegalMaster:
    def __init__(self):
        self.articulos = {}  # Diccionario principal { "100": "Texto..." }

    def limpiar_texto_boe(self, texto):
        """Elimina cabeceras, pies de página y une líneas cortadas."""
        lineas = texto.split('\n')
        lineas_limpias = []
        for linea in lineas:
            # Filtro de basura recurrente en PDFs del BOE
            if any(x in linea for x in ["BOLETÍN OFICIAL", "LEGISLACIÓN CONSOLIDADA", "Página ", "Referencia: BOE"]):
                continue
            lineas_limpias.append(linea)
        return "\n".join(lineas_limpias)

    # -------------------------------------------------------------------------
    # PASO 1: CARGAR LA LEY BASE (2020)
    # -------------------------------------------------------------------------
    def cargar_original(self):
        print(f"--- 1. Leyendo Texto Refundido 2020: {PDF_ORIGINAL} ---")
        texto_completo = ""
        try:
            with pdfplumber.open(PDF_ORIGINAL) as pdf:
                for page in pdf.pages:
                    texto_completo += page.extract_text() + "\n"
        except FileNotFoundError:
            print(f"ERROR: No encuentro el archivo {PDF_ORIGINAL}")
            return

        texto_limpio = self.limpiar_texto_boe(texto_completo)
        
        # Regex para artículos originales (formato estándar sin comillas)
        patron = re.compile(r'(?:^|\n)\s*Art[íi]culo\s+(\d+)\.?(.*?)((?=\n\s*Art[íi]culo)|\Z)', re.DOTALL | re.IGNORECASE)
        
        for match in patron.finditer(texto_limpio):
            num = match.group(1)
            contenido = f"Artículo {num}.{match.group(2)}"
            self.articulos[num] = contenido.strip()
        
        print(f"   -> Artículos base cargados: {len(self.articulos)}")

    # -------------------------------------------------------------------------
    # PASO 2: APLICAR LA REFORMA (2022) CON LÓGICA AVANZADA
    # -------------------------------------------------------------------------
    def aplicar_reforma(self):
        print(f"--- 2. Procesando Reforma 2022: {PDF_REFORMA} ---")
        texto_reforma = ""
        try:
            with pdfplumber.open(PDF_REFORMA) as pdf:
                for page in pdf.pages:
                    texto_reforma += page.extract_text() + "\n"
        except FileNotFoundError:
            print(f"ERROR: No encuentro el archivo {PDF_REFORMA}")
            return
        
        texto_reforma = self.limpiar_texto_boe(texto_reforma)

        # Dividimos el texto usando las INSTRUCCIONES como separadores.
        # Esto nos permite saber el contexto (de qué artículo se está hablando).
        regex_instruccion = r'(Se (?:modifica|añade|introduce|incorpora|da nueva redacción).*?Art[íi]culo\s+(\d+))'
        
        bloques = re.split(regex_instruccion, texto_reforma, flags=re.IGNORECASE)
        
        # Iteramos saltando de 3 en 3 (Texto previo, Instrucción capturada, Número Art capturado)
        i = 1
        while i < len(bloques) - 1:
            instruccion = bloques[i]      # Ej: "Se modifica el apartado 2 del artículo 58"
            num_art = bloques[i+1]        # Ej: "58"
            contenido_bloque = bloques[i+2] # El texto que viene a continuación
            
            # --- LÓGICA DE DECISIÓN ---
            
            # A. DETECTAR SUPRESIONES
            if any(x in instruccion.lower() for x in ["suprime", "deroga", "deja sin contenido"]):
                # ¿Es parcial (apartado) o total?
                if any(x in instruccion.lower() for x in ["apartado", "número", "letra", "punto"]):
                    # Intento de extracción del número de apartado a borrar
                    match_ap = re.search(r'(?:apartado|número|punto)\s+(\d+|[a-z]\))', instruccion, re.IGNORECASE)
                    if match_ap and num_art in self.articulos:
                        ap_borrar = match_ap.group(1)
                        self.borrar_apartado(num_art, ap_borrar)
                        print(f"   [-] Art. {num_art}: Suprimido Apartado {ap_borrar}")
                else:
                    # Supresión total
                    if num_art in self.articulos:
                        del self.articulos[num_art]
                        print(f"   [X] Art. {num_art}: ELIMINADO COMPLETAMENTE")

            # B. DETECTAR MODIFICACIONES / ADICIONES (Busca comillas «...»)
            else:
                # Regex estricto: Solo captura lo que esté dentro de comillas BOE
                regex_comillas = r'(?:«|<<|“|")\s*(.*?)(?:»|>>|”|")'
                match_contenido = re.search(regex_comillas, contenido_bloque, re.DOTALL)
                
                if match_contenido:
                    texto_nuevo = match_contenido.group(1).strip()
                    
                    # CASO 1: Es un Artículo Completo (Empieza por "Artículo X.")
                    if re.match(r'^Art[íi]culo\s+\d+', texto_nuevo, re.IGNORECASE):
                        self.articulos[num_art] = texto_nuevo
                        print(f"   [O] Art. {num_art}: Reemplazo COMPLETO")
                    
                    # CASO 2: Es un Apartado Suelto (Empieza por "4." o "a)")
                    elif re.match(r'^(\d+|[a-z]\))[\.\-\)\s]', texto_nuevo):
                        # Extraemos el número del apartado
                        match_num = re.match(r'^(\d+|[a-z]\))', texto_nuevo)
                        num_ap = match_num.group(1)
                        
                        if num_art in self.articulos:
                            self.inyectar_apartado(num_art, num_ap, texto_nuevo)
                            print(f"   [>] Art. {num_art}: Modificado Apartado {num_ap}")
                        else:
                            # Si el artículo no existe y nos dan un apartado, lo creamos
                            self.articulos[num_art] = f"Artículo {num_art}.\n{texto_nuevo}"
                            print(f"   [+] Art. {num_art}: Creado nuevo (desde apartado)")
                    
                    # CASO 3: Fallback (Asumimos reemplazo completo si hay duda y es "Se añade")
                    elif "se añade" in instruccion.lower() and num_art not in self.articulos:
                        self.articulos[num_art] = f"Artículo {num_art}. {texto_nuevo}"
                        print(f"   [+] Art. {num_art}: Añadido nuevo")

            i += 3 # Avanzamos al siguiente bloque

    # --- AYUDANTES DE MODIFICACIÓN DE TEXTO ---
    def borrar_apartado(self, num_art, num_apartado):
        texto = self.articulos[num_art]
        # Regex busca desde el número del apartado hasta el siguiente número
        regex = fr"(\n\s*{re.escape(num_apartado)}[\.\-\)].*?)(?=\n\s*(?:\d+|[a-z])[\.\-\)]|\Z)"
        # Reemplazamos por marca de supresión
        self.articulos[num_art] = re.sub(regex, f"\n{num_apartado}. (Suprimido)", texto, flags=re.DOTALL)

    def inyectar_apartado(self, num_art, num_apartado, texto_nuevo):
        texto = self.articulos[num_art]
        # Regex busca dónde está el apartado actualmente
        regex = fr"(\n\s*{re.escape(num_apartado)}[\.\-\)].*?)(?=\n\s*(?:\d+|[a-z])[\.\-\)]|\Z)"
        
        if re.search(regex, texto):
            # Si existe, lo sustituimos
            self.articulos[num_art] = re.sub(regex, f"\n{texto_nuevo}", texto, count=1, flags=re.DOTALL)
        else:
            # Si no existe, lo añadimos al final
            self.articulos[num_art] += f"\n{texto_nuevo}"

    # -------------------------------------------------------------------------
    # PASO 3: GENERAR PDF ORDENADO (LATÍN INCLUIDO)
    # -------------------------------------------------------------------------
    def generar_pdf(self):
        print(f"--- 3. Generando PDF Final: {PDF_SALIDA} ---")
        pdf = FPDF()
        pdf.set_auto_page_break(auto=True, margin=15)
        pdf.add_page()
        
        # Título del Documento
        pdf.set_font("Arial", "B", 16)
        pdf.cell(0, 10, "LEY CONCURSAL CONSOLIDADA", ln=True, align='C')
        pdf.set_font("Arial", "I", 12)
        pdf.cell(0, 10, "(Actualizada con Reforma 2022)", ln=True, align='C')
        pdf.ln(10)

        # --- ORDENACIÓN MAESTRA (Bis, Ter, Quater...) ---
        def obtener_valor_orden(clave):
            # Extraer número base
            match = re.match(r"(\d+)", clave)
            if not match: return 999999 # Al fondo si no tiene número
            
            numero = int(match.group(1))
            decimal = 0.0
            clave_low = clave.lower()

            # Tabla de sufijos latinos
            sufijos = {
                "bis": 0.01, "ter": 0.02, "quater": 0.03, "quinquies": 0.04,
                "sexies": 0.05, "septies": 0.06, "octies": 0.07, "nonies": 0.08,
                "decies": 0.09, "undecies": 0.10, "duodecies": 0.11
            }
            
            for sufijo, valor in sufijos.items():
                if sufijo in clave_low:
                    decimal = valor
                    break
            
            return numero + decimal

        # Ordenar las claves
        claves_ordenadas = sorted(self.articulos.keys(), key=obtener_valor_orden)

        # Escribir en el PDF
        for num in claves_ordenadas:
            texto = self.articulos[num]
            
            # Limpieza de codificación para FPDF (Latin-1)
            try:
                texto_safe = texto.encode('latin-1', 'replace').decode('latin-1')
            except:
                continue 

            # Separar Título (Primera frase) del Cuerpo
            partes = texto_safe.split('.', 1)
            titulo = partes[0] + "."
            cuerpo = partes[1] if len(partes) > 1 else ""

            # Estilo Título
            pdf.set_font("Arial", "B", 11)
            pdf.multi_cell(0, 6, titulo, align='J')
            
            # Estilo Cuerpo
            if cuerpo.strip():
                pdf.set_font("Arial", "", 10)
                pdf.multi_cell(0, 5, cuerpo.strip(), align='J')
            
            pdf.ln(4) # Espacio entre artículos

        pdf.output(PDF_SALIDA)
        print("¡PROCESO FINALIZADO CON ÉXITO!")

# --- EJECUCIÓN PRINCIPAL ---
if __name__ == "__main__":
    app = ConsolidadorLegalMaster()
    app.cargar_original()
    app.aplicar_reforma()
    app.generar_pdf()
```
