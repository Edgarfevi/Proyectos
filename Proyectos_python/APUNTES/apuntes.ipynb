# CONFIG KERNELLS JUPYTER, ENTORNOS VIRTUALES DE PYTHON:
Es neceesario hacer dentro del entorno: pip install jupyter ipykernel
Para la configuración el siguiente comando: python3.12 -m ipykernel install --prefix ~/.local --name (nombre interno) --display-name "(nombre que visualizamos)"
No usamos este comando: python3.12 -m ipykernel install --user --name electronica --display-name "Python (Electronica)"
El segundo no nos detectará bien el kernell
