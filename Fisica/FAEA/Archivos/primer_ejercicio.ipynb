# ***Resumen***

Para este ejercicio trataremos de representar la energía faltante en el plano trasverso, con ciertas restricciones a partir del árbol ubicado en el archvivo ttbar.root

# ***Procedimiento***

Hemos querido representar la energía faltante en el plano trasverso, esta variable no existe dentro de nuestro árbol, luego la pintaremos a partir de los datos si conocidos en el archivo, pudiendo así acceder a la variables MET_px y MET_py necesarias para determinar la energía en el plano trasverso.

Este archivo en concreto coincide con un choque proto-proton, dode se porduce el par quark top y antitop.

Las restricciones que ponemos son en esencia si se ha producido o no 1 muon exacto.

Para hablar de lo más imoportante, en este caso los histogramas, haremos mención a que representamos en los ejes, energía en el x, en el y representamos cantidad de eventos, escalado, ya que en este caso nos importa que porcentaje de sucesos dan un cierto de energía y a la hora de suporponer es esencialmente más útil para comparar que la cantidad de sucesos.

Además tras ciertas visualizaciones observamos que hay energías que no se dan, luego ajustamos los limites de nuestra representación para que se ajuste lo más posible a nuestros datos, y seleccionamos en este caso 30 bines, lo cual me parece justo para observar rangos de enrgía relativamente cercanos, y un error asumible para la representación, cuanto más aumentamos el numero de bines asumimos que medimos con muchisima más precisión por eso he escogido 30. Además si escogiesemos menos no observariamos la tendencia real de la variable.

# ***Representación y código***

```python
import ROOT as rt #Libreria/framewoork usado

f = rt.TFile("../Datos/Introducción_root/ttbar.root") #abrir el archivo .root
t = f.Get("events")  # Obtener el árbol directamente

# Crear un lienzo (canvas) de 800x600 píxeles para dibujar los histogramas
new_canvas = rt.TCanvas("new_canvas", "Canvas de histograma", 800, 600)

# Crear primer histograma (h) para eventos con 0 muones
# 30 bins, rango de 0 a 400 GeV
h = rt.TH1F("EFPT0", "Energía Faltante plano trasverso", 30, 0, 400)
h.SetXTitle("MET (GeV)")  # Etiqueta del eje X
h.SetYTitle("cantidad de eventos")  # Etiqueta del eje Y
h.SetLineColor(rt.kBlue)  # Color de línea azul
h.SetLineWidth(2)  # Grosor de línea

h.SetStats(0)  # Ocultar estadísticas
# Crear leyenda en posición (0.6,0.9) a (0.9,0.8)
legend = rt.TLegend(0.6,0.9,0.9,0.8)  # x1,y1,x2,y2
legend.SetHeader("Legend","C")
legend.AddEntry(h,"Histograma para 0 muones","l")

# Crear segundo histograma (h2) para eventos con 1+ muones
h2 = rt.TH1F("EFPT1", "Energia Faltante plano trasverso", 30, 0, 400)
h2.SetXTitle("MET (GeV)")
h2.SetYTitle("cantidad de eventos")
h2.SetLineColor(rt.kRed)  # Color de línea rojo
h2.SetLineWidth(2)
h2.SetStats(0)  # Ocultar estadísticas

# Añadir h2 a la leyenda
legend.SetHeader("Legend","C")
legend.AddEntry(h2,"Histograma para 1 muones","l")

# Llenar los histogramas desde el árbol (t)
# MET = raíz cuadrada de (px² + py²)
# "goff" = no mostrar (solo rellenar)
t.Draw("sqrt(MET_px*MET_px + MET_py*MET_py)>>EFPT1","NMuon == 0","goff")  # h2: eventos con 0 muones
t.Draw("sqrt(MET_px*MET_px + MET_py*MET_py)>>EFPT0","NMuon == 1","goff")  # h: solo eventos con 1 muones


# Normalizar ambos histogramas dividiendo por su integral (área total)
# Esto permite comparar formas con diferentes números de eventos
h.Scale(1.0/h.Integral())
h2.Scale(1.0/h2.Integral())

# Superponer h2 en el mismo gráfico
h2.Draw("hist")
# Dibujar h como histograma de barras
h.Draw("hist same")

# Mostrar la leyenda
legend.Draw()

# Mostrar el lienzo y guardar como imagen
new_canvas.Draw()
new_canvas.Print("histograma_comparativo.png")  # Exportar a PNG
```
