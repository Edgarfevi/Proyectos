# Estudio del bosón Z en canales dileptónicos ($ee$ y $\mu\mu$)

Este cuaderno organiza un flujo completo de análisis para eventos de física de partículas en tres niveles: **generación (gen)**, **reconstrucción (reco)** y **datos (data)**. El objetivo es comparar la cinemática de leptones y la distribución de masa invariante para evaluar resolución instrumental, sesgos de reconstrucción y compatibilidad entre simulación y datos.

## Objetivos físicos y computacionales

1. Construir variables cinemáticas por evento ($p_x, p_y, p_z, p_T, \eta, \phi, E$).
2. Calcular la masa invariante del sistema dileptónico para identificar el pico del bosón $Z$.
3. Visualizar y comparar distribuciones entre niveles de procesamiento.
4. Ajustar el pico con distintos modelos (Gauss, Breit-Wigner, Voigt y Cruijff).
5. Realizar comparativas globales entre canales y tipos de muestra.

## Ecuaciones clave

A partir de dos cuatro-momentos $p_1$ y $p_2$:

$$
p_{\ell\ell} = p_1 + p_2
$$

$$
m_{\ell\ell}^2 = p_{\ell\ell}^2 = E_{\ell\ell}^2 - |\vec{p}_{\ell\ell}|^2
$$

También se emplean variables de uso estándar:

$$
p_T = \sqrt{p_x^2 + p_y^2}, \qquad
\eta = -\ln\left[\tan\left(\frac{\theta}{2}\right)\right]
$$

#### Ajuste del pico de masa invariante

Se aplican funciones de ajuste para caracterizar la forma del pico del bosón $Z$. Conceptualmente:
- **Gaussiana**: aproxima resolución efectiva cuando domina el ensanchamiento instrumental.
- **Breit-Wigner**: modela anchura natural de resonancia.
- **Voigt**: convolución de anchura física e instrumental.
- **Cruijff**: forma empírica con colas asimétricas.

El diagnóstico principal se interpreta con $\chi^2/\mathrm{NDF}$ y estabilidad de los parámetros del pico.

```python

Electron_gen.plot_mass_fits()
```
