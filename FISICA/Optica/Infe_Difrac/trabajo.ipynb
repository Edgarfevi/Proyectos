# Interferencias múltiples y difracción de una apertura pentagonal

## Introducción
Los fenómenos de interferencia en medios estratificados y de difracción por aperturas finitas constituyen dos pilares de la óptica ondulatoria. En este informe se estudian ambos problemas dentro del marco escalar y monocromático, con el objetivo de relacionar directamente las predicciones teóricas con resultados numéricos.

En la primera parte se analiza una multicapa dieléctrica periódica formada por la repetición de bicapas con índices de refracción $n_1=1.49$ y $n_2=1.73$. Se evalúa la reflectancia en función del número de bicapas y de la longitud de onda, verificando las condiciones de fase que producen máximos y mínimos de reflexión.

En la segunda parte se determina el patrón de difracción de Fraunhofer de una apertura pentagonal regular en el plano de transferencia de momento $(q_y,q_z)$, incluyendo el análisis bidimensional de intensidad y su corte central.

## Parámetros del estudio
- Medio incidente y de salida: $n_0=n_{sub}=1$.
- Longitud de onda de referencia para diseño de multicapa: $\lambda_0=500\,\mathrm{nm}$.
- Rango espectral analizado: $450$-$650\,\mathrm{nm}$.
- Apertura de difracción: pentágono regular con radio característico $R=75\,\mu\mathrm{m}$.

# Metodología y marco teórico

## Interferencias múltiples en multicapas
La reflectancia total de una estructura estratificada se obtiene a partir de la superposición coherente de las reflexiones parciales en cada intercara. Para incidencia normal, la fase de propagación en una capa de índice $n$ y espesor $d$ es
$$
\phi=\frac{2\pi n d}{\lambda}.
$$

El cálculo global se formula mediante matrices de transferencia de cada capa y su producto ordenado a lo largo del apilamiento. A partir de la matriz total se obtiene el coeficiente complejo de reflexión $r$, y la reflectancia se evalúa como
$$
R=|r|^2.
$$

Se estudian específicamente dos familias de diseño:
- Máximos de reflectancia: $\phi=(2m+1)\pi$.
- Mínimos de reflectancia: $\phi=2m\pi$.

Estas condiciones equivalen, respectivamente, a espesores tipo cuarto de onda generalizado y múltiplos de media onda.

## Difracción de Fraunhofer en apertura pentagonal
En el régimen de campo lejano, el campo difractado se expresa como transformada de Fourier de la función apertura:
$$
E(q_y,q_z)\propto\iint_A e^{-i(q_y y+q_z z)}\,dy\,dz,
$$
con intensidad
$$
I(q_y,q_z)=|E(q_y,q_z)|^2.
$$

Para preservar de forma explícita la geometría del pentágono, el dominio se descompone en cinco triángulos equivalentes. El campo total se obtiene como suma coherente de cinco contribuciones rotadas:
$$
E(q_y,q_z)=\sum_{j=1}^{5}E_j(q_y,q_z),
\qquad
\alpha_j=\frac{2\pi(j-1)}{5}.
$$

### Desarrollo de la integral de Fraunhofer

Sea $\gamma=\pi/5$ y $R$ el radio del pentágono regular. Para un triángulo de referencia, una parametrización conveniente de la región es:

$$-R\cos\gamma\le z\le 0, \qquad z\tan\gamma\le y\le -z\tan\gamma.$$

Entonces, el campo difractado por este triángulo es:

$$E_1(q_y,q_z)=\int_{-R\cos\gamma}^{0}\int_{z\tan\gamma}^{-z\tan\gamma} e^{-i(q_y y+q_z z)}dydz.$$

Separando la dependencia en $y$:

$$E_1=\int_{-R\cos\gamma}^{0}e^{-iq_z z}\left[\int_{z\tan\gamma}^{-z\tan\gamma}e^{-iq_y y}dy\right]dz.$$

Resolviendo la integral interior se obtiene:

$$\int_{z\tan\gamma}^{-z\tan\gamma}e^{-iq_y y}dy = \left[\frac{e^{-iq_y y}}{-iq_y}\right]_{z\tan\gamma}^{-z\tan\gamma} = \frac{e^{iq_y z\tan\gamma}-e^{-iq_y z\tan\gamma}}{-iq_y} = \frac{e^{-iq_y z\tan\gamma}-e^{iq_y z\tan\gamma}}{iq_y}.$$

Sustituyendo en la integral principal y multiplicando las exponenciales para agrupar la dependencia en la variable $z$:

$$E_1=\frac{1}{iq_y}\int_{-R\cos\gamma}^{0}\left[e^{-i(q_z+q_y\tan\gamma)z}-e^{-i(q_z-q_y\tan\gamma)z}\right]dz.$$

Definiendo las constantes para simplificar la notación:

$$a_+=q_z+q_y\tan\gamma, \qquad a_-=q_z-q_y\tan\gamma,$$

La expresión se reduce a dos integrales sencillas:

$$E_1=\frac{1}{iq_y}\left[\int_{-R\cos\gamma}^{0}e^{-ia_+z}dz-\int_{-R\cos\gamma}^{0}e^{-ia_-z}dz\right].$$

Resolviendo la integral genérica en $z$:

$$\int_{-R\cos\gamma}^{0}e^{-iaz}dz = \left[\frac{e^{-iaz}}{-ia}\right]_{-R\cos\gamma}^{0} = \frac{1-e^{iaR\cos\gamma}}{-ia} = \frac{e^{iaR\cos\gamma}-1}{ia}.$$

Sustituyendo este resultado, se obtiene la expresión analítica final para la amplitud del campo de un único triángulo:

$$E_1(q_y,q_z)=\frac{1}{iq_y}\left[\frac{e^{i(q_z+q_y\tan\gamma)R\cos\gamma}-1}{i(q_z+q_y\tan\gamma)} - \frac{e^{i(q_z-q_y\tan\gamma)R\cos\gamma}-1}{i(q_z-q_y\tan\gamma)}\right].$$

Finalmente, para considerar el pentágono completo, a cada uno de los 5 triángulos rotados se le aplican las componentes transformadas del vector $\vec{q}$:

$$q_{y,j}=q_y\cos\alpha_j+q_z\sin\alpha_j, \qquad q_{z,j}=-q_y\sin\alpha_j+q_z\cos\alpha_j,$$

donde $\alpha_j=\frac{2\pi(j-1)}{5}$ para $j=1,...,5$. Se evalúa la contribución $E_j=E_1(q_{y,j},q_{z,j})$, y se suma de forma coherente para obtener el campo total y la intensidad de difracción $I(q_y, q_z) = \left|\sum_{j=1}^{5} E_j\right|^2$.

```python
# Imports y configuración de trazado
# Numpy para cálculos numéricos, matplotlib para gráficos.
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.path import Path
from matplotlib.colors import LogNorm
from matplotlib.ticker import ScalarFormatter

# Ajustes globales de estilo para las figuras
plt.rcParams.update({
    "figure.figsize": (9, 5),
    "axes.grid": True,
    "grid.alpha": 0.25,
    "font.size": 11,
    "axes.spines.top": False,
    "axes.spines.right": False,
})

# Parámetros físicos: índices y longitud de onda de referencia
n0 = 1.0  # medio incidente
n_sub = 1.0  # medio de sustrato / salida
n1 = 1.49
n2 = 1.73
lambda0 = 500e-9  # longitud de onda de diseño (m)


# --- Funciones físicas ---
# Matriz de propagación de una capa (incidencia normal)
# Devuelve la matriz 2x2 compleja que representa la capa
def layer_matrix(n, d, wavelength):
    delta = 2 * np.pi * n * d / wavelength
    return np.array(
        [[np.cos(delta), 1j * np.sin(delta) / n],
         [1j * n * np.sin(delta), np.cos(delta)]],
        dtype=complex,
    )


# Cálculo de la reflectancia de un apilamiento a partir
# del producto de matrices de transferencia y las condiciones de
# frontera (n_incident, n_exit)
def stack_reflectance(thicknesses, indices, wavelength, n_incident=1.0, n_exit=1.0):
    # Producto ordenado de las matrices de cada capa
    matrix = np.eye(2, dtype=complex)
    for n_layer, d_layer in zip(indices, thicknesses):
        matrix = matrix @ layer_matrix(n_layer, d_layer, wavelength)

    # Coeficiente complejo de reflexión a partir de elementos de la matriz total
    q0 = n_incident
    qs = n_exit
    numerator = q0 * matrix[0, 0] + q0 * qs * matrix[0, 1] - matrix[1, 0] - qs * matrix[1, 1]
    denominator = q0 * matrix[0, 0] + q0 * qs * matrix[0, 1] + matrix[1, 0] + qs * matrix[1, 1]
    r = numerator / denominator
    return r, abs(r) ** 2


# Cálculo de los espesores de una bicapa para los casos pedidos
# mode = "max" (φ=(2m+1)π) o "min" (φ=2mπ)
def bilayer_thicknesses(wavelength, n_a, n_b, m, mode):
    if mode == "max":
        phase = (2 * m + 1) * np.pi
        d_a = (2 * m + 1) * wavelength / (4 * n_a)
        d_b = (2 * m + 1) * wavelength / (4 * n_b)
    elif mode == "min":
        phase = 2 * np.pi * m
        d_a = m * wavelength / (2 * n_a)
        d_b = m * wavelength / (2 * n_b)
    else:
        raise ValueError("mode debe ser 'max' o 'min'")
    return d_a, d_b, phase
```

```python
# Bloque: cuatro gráficas de reflectancia frente al número de bicapas
# Se generan cuatro casos pedidos en el enunciado y se grafican
cases = [
    ("máximo", 0, "m=0", "max"),
    ("máximo", 2, "m=2", "max"),
    ("mínimo", 1, "m=1", "min"),
    ("mínimo", 2, "m=2", "min"),
]

num_bilayers = np.arange(1, 26)
indices = [n1, n2]

fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=True, sharey=True)
axes = axes.ravel()

for ax, (kind, m_value, label, mode) in zip(axes, cases):
    # Calcula espesores según el modo y m pedido
    d1, d2, phase = bilayer_thicknesses(lambda0, n1, n2, m_value, mode)
    thicknesses = [d1, d2]
    reflectances = []
    for n_pairs in num_bilayers:
        # Para cada número de pares se calcula la reflectancia
        r, R = stack_reflectance(thicknesses * n_pairs, indices * n_pairs, lambda0, n0, n_sub)
        reflectances.append(R)

    reflectances = np.array(reflectances)
    ax.plot(num_bilayers, reflectances,"o-", color="#1f77b4", lw=4)
    ax.set_title(f"{kind} - {label}")
    ax.set_xlabel("Número de bicapas")
    ax.set_ylabel("Reflectancia")
    ax.text(
        0.75,
        0.12,
        f"d1 = {d1*1e9:.3f} nm\nd2 = {d2*1e9:.3f} nm\nphi = {phase/np.pi:.0f}π",
        transform=ax.transAxes,
        fontsize=9,
        bbox=dict(boxstyle="round,pad=0.35", facecolor="white", alpha=0.8),
    )

    # Impresión de resultados para mantener salida textual en el informe
    print(f"{kind} {label}: d1 = {d1*1e9:.3f} nm, d2 = {d2*1e9:.3f} nm, fase esperada = {phase/np.pi:.0f}π")

fig.suptitle("Reflectancia de la multicapa en función del número de bicapas", y=1.02, fontsize=14)
fig.tight_layout()
plt.savefig("reflectancia_multicapa.png", dpi=300)
plt.show()
```

```python
# Barrido espectral R(λ) para el caso diseñado en 500 nm
m_max = 0
d1_max, d2_max, phase_max = bilayer_thicknesses(lambda0, n1, n2, m_max, "max")
num_pairs_sweep = 25
wavelengths = np.linspace(450e-9, 650e-9, 600)
R_wavelength = []

# Se calcula la reflectancia para cada longitud de onda
for wavelength in wavelengths:
    r, R = stack_reflectance(
        [d1_max, d2_max] * num_pairs_sweep,
        [n1, n2] * num_pairs_sweep,
        wavelength,
        n0,
        n_sub,
    )
    R_wavelength.append(R)

R_wavelength = np.array(R_wavelength)

# Gráfico de la dependencia espectral
plt.figure(figsize=(9, 5))
plt.plot(wavelengths * 1e9, R_wavelength, color="#d62728", lw=2)
plt.axvline(500, color="0.35", ls="--", lw=1)
plt.scatter([500], [stack_reflectance([d1_max, d2_max] * num_pairs_sweep, [n1, n2] * num_pairs_sweep, lambda0, n0, n_sub)[1]], color="#1f77b4", zorder=3)
plt.xlabel("Longitud de onda (nm)")
plt.ylabel("Reflectancia")
plt.title("Dependencia espectral para la multicapa de máxima reflectancia en 500 nm")
plt.ylim(0, 1.02)
plt.tight_layout()
plt.savefig("reflectancia_multicapa_500nm.png", dpi=300)
plt.show()

# Salidas impresas para incluir valores clave en el informe
print(f"Caso usado: m=0, d1 = {d1_max*1e9:.3f} nm, d2 = {d2_max*1e9:.3f} nm, fase esperada = {phase_max/np.pi:.0f}π")
print(f"Reflectancia en 500 nm con {num_pairs_sweep} bicapas: {stack_reflectance([d1_max, d2_max] * num_pairs_sweep, [n1, n2] * num_pairs_sweep, lambda0, n0, n_sub)[1]:.6f}")
```

# Resultados y discusión: multicapas

Los resultados obtenidos para $R(N)$, con $N$ número de bicapas, muestran dos comportamientos claramente diferenciados:

- En los casos con fase impar de $\pi$, la reflectancia crece de manera monótona con $N$ y alcanza valores altos, característicos de un reflector dieléctrico de Bragg.
- En los casos con fase múltiplo de $2\pi$, la reflectancia permanece muy reducida, coherente con cancelación interferencial en reflexión dentro del modelo ideal sin pérdidas.

Desde un punto de vista físico, el incremento de contraste en reflexión con el número de periodos se debe al refuerzo progresivo de la interferencia constructiva para las longitudes de onda próximas a la de diseño.

En el análisis espectral $R(\lambda)$ entre $450$ y $650\,\mathrm{nm}$ se observa un máximo bien localizado alrededor de $500\,\mathrm{nm}$, confirmando la selectividad espectral del sistema. Este resultado valida la elección de espesores y la condición de fase impuesta en el diseño de la bicapa.

```python
# Función etri: amplitud del campo difractado por un triángulo de referencia
# Implementa la expresión analítica derivada en el informe.
def etri(qy, qz, R):
    """Calcula el campo difractado por uno de los triángulos (Ec. 17 del PDF)."""
    gamma = np.pi / 5
    L = R * np.cos(gamma)
    tol = 1e-6 
    
    # Evitar división por cero en qy
    qy_safe = np.where(np.abs(qy) < tol, tol, qy)
    a_minus = qz - qy_safe * np.tan(gamma)
    a_plus  = qz + qy_safe * np.tan(gamma)
    
    # Integral genérica en z evaluada analíticamente (tratando a≈0)
    def f_int(a):
        res = np.empty_like(a, dtype=complex)
        mask = np.abs(a) < tol
        # Si a≈0, la integral en z da L (longitud del segmento)
        res[mask] = L
        # En caso general, usar la forma cerrada (1 - e^{-i a L})/(i a)
        res[~mask] = (1 - np.exp(-1j * a[~mask] * L)) / (1j * a[~mask])
        return res
    
    # La expresión final proviene de la integral en y seguida de la integral en z
    return (1 / (1j * qy_safe)) * (f_int(a_minus) - f_int(a_plus))


# Función que construye la intensidad total sumando las 5 contribuciones rotadas
def difraccion_pentagono_analitica(R=75e-6, qmax=5e5, n=1200):
    """Calcula la intensidad analítica en un rango mayor de q."""
    qy = np.linspace(-qmax, qmax, n)
    qz = np.linspace(-qmax, qmax, n)
    QY, QZ = np.meshgrid(qy, qz)
    
    E_total = np.zeros_like(QY, dtype=complex)
    for j in range(1, 6):
        # Rotación de las componentes de q por el ángulo α_j
        alpha = 2 * np.pi * (j - 1) / 5
        qy_j = QY * np.cos(alpha) + QZ * np.sin(alpha)
        qz_j = -QY * np.sin(alpha) + QZ * np.cos(alpha)
        E_total += etri(qy_j, qz_j, R)
        
    intensidad = np.abs(E_total)**2
    intensidad /= intensidad.max()
    return qy, qz, intensidad

# --- Parámetros y Cálculo ---
q_limite = 5e5  # Espacio aumentado
qy, qz, intensidad = difraccion_pentagono_analitica(R=75e-6, qmax=q_limite, n=1200)

# --- Configuración de las Gráficas ---
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# 1. Mapa de Intensidad 2D
im = ax[0].imshow(intensidad, origin="lower", 
                  extent=(qy[0], qy[-1], qz[0], qz[-1]), 
                  cmap="magma", norm=LogNorm(vmin=1e-6, vmax=1))
ax[0].set_title("Patrón de Difracción")
ax[0].set_xlabel(r"$q_y$ (rad/m)")
ax[0].set_ylabel(r"$q_z$ (rad/m)")
fig.colorbar(im, ax=ax[0], fraction=0.046, pad=0.04)

# 2. Corte Central 1D: perfil en q_z = 0
centro = len(qz) // 2
linea = intensidad[centro, :]
ax[1].plot(qy, linea, lw=1.5, color="darkblue")
ax[1].set_title("Perfil de Intensidad en $q_z=0$")
ax[1].set_xlabel(r"$q_y$ (rad/m)")
ax[1].set_ylabel("Intensidad normalizada")
ax[1].set_ylim(-0.02, 1.05)
ax[1].grid(alpha=0.3)

# Formateo de ejes para notación científica cuando procede
for a in ax:
    formatter = ScalarFormatter(useMathText=True)
    formatter.set_scientific(True)
    formatter.set_powerlimits((-1,1)) 
    a.xaxis.set_major_formatter(formatter)
    if a == ax[0]:
        a.yaxis.set_major_formatter(formatter)

# Marcar primeros mínimos locales en el corte central (para el informe)
minimos_idx = []
for i in range(centro + 1, len(qy) - 1):
    if linea[i] < linea[i-1] and linea[i] < linea[i+1]:
        minimos_idx.append(i)
        if len(minimos_idx) == 5: 
            break

for idx in minimos_idx:
    ax[1].plot(qy[idx], linea[idx], "ro", ms=4)

plt.tight_layout()
plt.savefig("difraccion_pentagono_analitica.png", dpi=300)
plt.show()
```
